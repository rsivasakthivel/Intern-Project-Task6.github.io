#### üëâ Import the Numpy and Pandas Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import tree
import sklearn.datasets as datasets
```

#### üëâ Load the iris dataset

```python
iris=datasets.load_iris()
```

#### üëâ Put all the features into X variable

```python
X=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
X.head()
```

#### üëâ Target values into y variable

```python
y=iris.target
y
```

#### üëâ Checking if there is any null value

```python
X.isnull().sum()
```

#### üëâ Describing the features of iris dataset

```python
X.describe()
```

#### üëâ Infromation about the features of iris dataset

```python
X.info()
```

#### üëâ Splitting the data into train and test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

#### üëâ  Creating an obeject for decision tree

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

#### üëâ Fitting the train dataset into the model

```python
model.fit(X_train,y_train)
```

#### üëâ Prediciting the test data

```python
y_predict=model.predict(X_test)
y_predict
```

#### üëâ  Visualizing the Decision Tree

```python
feature_n=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
class_n=['setosa','versicolor','virginica']
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)
tree.plot_tree(model,feature_names=feature_n,class_names=class_n,filled='True',rounded='True')
```

#### üëâ Accuracy of the model

```python
model.score(X_test,y_test)
```
